```python
# Dependencies and Setup
import pandas as pd
import os


# File to Load (Remember to Change These)
pwd = os.path.abspath('.')

file_to_load = os.path.join(pwd, 'car_data.csv')

# Read Purchasing File and store into Pandas data frame
car_data = pd.read_csv(file_to_load)
car_data.head()

```

```python
#Source: Homework Solution for Mice Data
# Find - values in Fast Charge Column
fc_value = car_data[car_data['FastCharge_KmH'] == '-']

fc_value


```

```python
# Drop - values in Fast Charge Column
clean_df = car_data[car_data['FastCharge_KmH'] != '-'] 
clean_df
```

```python
#check for Null values
#https://stackoverflow.com/questions/26266362/how-to-count-the-nan-values-in-a-column-in-pandas-dataframe/26266451#26266451
print(car_data.isnull().sum())
```

```python
#drop unneeded columns
#source: https://pandas.pydata.org/pandas-docs/version/0.21/generated/pandas.DataFrame.drop.html
new_df = clean_df.drop(['RapidCharge', 'PowerTrain', 'PlugType', 'Segment', 'PriceEuro'], axis = 1)
new_df.head()
```

```python
#export csv file
#source https://datatofish.com/export-dataframe-to-csv/
new_df.to_csv(r'C:\Users\mom\Documents\GitHub\project_2_etl\clean_car_data.csv')
```
